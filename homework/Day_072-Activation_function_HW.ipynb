# Rectified Linear Unit- Relu 

f(x)=max(0,x)


```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline

def Relu(x):
    return(np.maximum(0,x))

def devRelu(x):
    return((Relu(x)>0).astype('int'))
```

```python
x=plt.linspace(-10,10,100)

plt.plot(x,Relu(x),'b')
plt.plot(x,devRelu(x),'r')

# 顯示後面網格
plt.grid()

plt.title('Relu Function')

# 顯現 the Sigmoid formula
# 前面兩數字是(x,y)
plt.text(2, 8.5, r'f(x)=max(0,x)', fontsize=12)
plt.text(4,1.2, r'derivative', fontsize=12)

plt.gca().xaxis.set_major_locator(plt.MultipleLocator(1))  #坐標軸的間距
plt.gca().yaxis.set_major_locator(plt.MultipleLocator(1))
```
