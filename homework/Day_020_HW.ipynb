```python
# 載入需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 另一個繪圖-樣式套件
plt.style.use('ggplot')

# 忽略警告訊息
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
np.random.randn(10) #返回標準常態的數字，可以給多個數字指定維度
# np.random.rand(n) 返回 0-1的數字
```

```python
"""
Your Code Here
"""
matrix = np.random.randn(10,10)
matrix=pd.DataFrame(matrix)
plt.figure(figsize=(10,10))
cormatrix=matrix.corr()
"""
Your Code Here
"""
#heatmap不是只能畫corr，也可以單純化資料
heatmap = sns.heatmap(data=cormatrix, cmap = plt.cm.RdYlBu_r, vmin = -0.25, annot = True, vmax = 0.6) # annot是否顯示資料值, vmax(min)顏色的上下界
                                                                                                    # cmap選擇matplot的顏色表
plt.show()
```

```python
nrow = 1000
ncol = 3
"""
Your Code Here
"""
matrix = pd.DataFrame(np.random.uniform(-1,1,(1000,3)))

# 隨機給予 0, 1, 2 三種標籤
#indice = np.random.choice([0,1,2], size=nrow) #隨機生成nrow個 0,1,2的數字
#plot_data = pd.DataFrame(matrix, indice).reset_index()

# 繪製 seborn 進階 Heatmap

#sns.pairgrid -> 生成pairgrid的物件
#vars就是有哪些變量想要拿來作圖    
#hue 就類似ggplot2 的 color
#size 圖像大小
grid = sns.PairGrid(data = matrix, size = 3, diag_sharey=True, 
                     vars = list(matrix.columns))
"""
Your Code Here
Please replace "..." to correct plot function
"""
grid.map_upper(plt.scatter , alpha = 0.2)
grid.map_diag(sns.distplot)
grid.map_lower(sns.kdeplot,alpha = 0.2)

plt.show()
```
