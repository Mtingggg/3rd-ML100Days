```python
# 載入需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 另一個繪圖-樣式套件

# 忽略警告訊息
%matplotlib inline
plt.style.use('ggplot')
import warnings
warnings.filterwarnings('ignore')

# 設定 data_path
dir_data = './data/'
```

```python
# 讀取檔案
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
app_train.head()
```

```python
#app_train['HOUSETYPE_MODE']=app_train['HOUSETYPE_MODE'].fillna("NA")
print(app_train['HOUSETYPE_MODE'].unique())
#app_train['AMT_INCOME_TOTAL']=np.log(app_train['AMT_INCOME_TOTAL']+1)

housetype=app_train['HOUSETYPE_MODE'].unique()
nrows = len(housetype)
ncols = nrows // 2

plt.figure(figsize=(10,30))
i=1
plt.subplots_adjust(hspace=0.4, wspace=0.4) #調整每張圖的間距

for house in housetype:
    plt.subplot(nrows,ncols,i)
    app_train.loc[(app_train['HOUSETYPE_MODE']==house)&(app_train['TARGET']==0),'AMT_INCOME_TOTAL'].hist()
    plt.title(house+" TARGET=0")
    i+=1
    plt.subplot(nrows,ncols,i)
    app_train.loc[(app_train['HOUSETYPE_MODE']==house)&(app_train['TARGET']==1),'AMT_INCOME_TOTAL'].hist()
    plt.title(house+" TARGET=1")
    i+=1
```
