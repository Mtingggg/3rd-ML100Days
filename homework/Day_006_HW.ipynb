```python
import os
import numpy as np
import pandas as pd
```

```python
# 設定 data_path, 並讀取 app_train
dir_data = './data/'
f_app_train = os.path.join(dir_data, 'application_train.csv')
app_train = pd.read_csv(f_app_train)

##./data/application_train.csv 
#代表程式會找的是跟 .ipython 相同目錄下 > data 資料夾 application_train.csv 檔案
# PS ./ 代表相同目錄、 ../ 代表上一層目錄
```

## 作業
將下列部分資料片段 sub_train 使用 One Hot encoding, 並觀察轉換前後的欄位數量 (使用 shape) 與欄位名稱 (使用 head) 變化

```python
sub_train = pd.DataFrame(app_train['WEEKDAY_APPR_PROCESS_START'])
print(sub_train.shape)
sub_train.head()
```

```python
sub_train=pd.get_dummies(sub_train)
print(sub_train.shape)
sub_train.head()
```

```python
#補充labelencoding兩種方法(此處星期應該用onehot encoding比較好)
#1 sklearn
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
le_count = 0

for col in sub_train:
    if len(sub_train[col])>2:
        #trian on training data
        le.fit(sub_train[col])
        #transform on training data
        sub_train[col]=le.transform(sub_train[col])
        
        le_count+=1
        
print('%d columns were label encoded' %le_count)
sub_train.head        
        
```

```python
#2 pandas
Mapp = {
    'WEDNESDAY':3,
    'MONDAY':1,
    'THURSDAY':4,
    'SUNDAY':7,
    'SATURDAY':6,
    'FRIDAY':5,
    'TUESDAY':2}
sub_train['WEEKDAY_APPR_PROCESS_START_LE']=sub_train['WEEKDAY_APPR_PROCESS_START'].map(Mapp)
sub_train.head()
```
